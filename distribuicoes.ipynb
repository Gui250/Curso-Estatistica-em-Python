## Distribuições

```python
import numpy as np
import matplotlib.pyplot as plt

# Parâmetros da simulação
np.random.seed(42) # Para reprodutibilidade
n_simulacoes = 1000 # Número de lançamentos

somas = []

for rodada in range(n_simulacoes):
    # Simular lançamentos de dois dados
    dado1 = np.random.randint(1, 7, 1)
    dado2 = np.random.randint(1, 7, 1)

    # Calcular a soma dos resultados
    soma = dado1 + dado2 
    somas.append(int(soma[0]))

# Visualização 
plt.hist(somas, bins=1)
plt.title("Distribuição da Soma de Dois Dados de 6 Faces")
plt.xlabel("Soma")
plt.ylabel("Frequência")
plt.xticks(range(2, 13)) # Valores possiveis para a soma de (2 a 12)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
