# Distribuição Normal

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parâmetros da distribuição normal
mu = 5 # Média
sigma = 1 # Distribuição

# Calcular a probabilidade de entrada em 6 dias ou menos
probabilidade = norm.cdf(6, loc=mu, scale=sigma)

```

```python
# Visualizacao da distribuicao 
x = np.linspace(mu - 4 * sigma, mu + 4*sigma, 100)
y = norm.pdf(x, loc=mu, scale=sigma)
```

```python
plt.plot(x, y, label="Distribuicao Normal")
plt.fill_between(x, y, where=x <= 6, color="blue", alpha=0.3, label="Entrega <= 6 dias")
plt.title("Tempo de entrega de pedidos")
plt.xlabel("Dias")
plt.ylabel("Probabilidade")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

print(f"Probabilidade de entrega em 6 dias ou menos: {probabilidade:.2%}")
```

# Distribuição Binomial

```python
from scipy.stats import binom
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt

# Parâmetros da distribuição binomial
n = 50 # Número de e-mails enviados
p = 0.2 # Probabilidade de abertura

# Calcular a probabilidade de exatamente 15 aberturas
k = 15
probabilidade = binom.pmf(k, n, p)
```

```python
probabilidade
```

```python
# Visualizacao da distribuicao
x = np.arange(0, n + 1)
y = binom.pmf(x, n, p)
```

```python
plt.bar(x, y, label="Distribuicao Binomial")
plt.axvline(k, color="red", linestyle="--", label="Numero de sucessos")
plt.title("Distribuicao Binomial")
plt.xlabel("Numero de aberturas")
plt.ylabel("Probabilidade")
plt.legend()
plt.show()
print(f"Probabilidade de exatamente 15 aberturas: {probabilidade:.2%}")
```

# Distribuição Poissonf

```python
from scipy.stats import poisson

# Parâmetros da distribuição
lambda_param = 10 # Taxa de pedidos por hora

# Calcular a probabilidade de exatamente 15 ocorrências
k = 15 
probabilidade = poisson.pmf(k, lambda_param)
```

```python
probabilidade
```

```python
x = np.arange(0, 30)
y = poisson.pmf(x, lambda_param)
```

```python
plt.bar(x, y, color="blue", alpha=0.6, label="Distribuicao Poisson")
plt.axvline(k, color="red", linestyle="--", label="Numero de ocorrências")
plt.title("Distribuição de Pedidos Por Hora")
plt.xlabel("Número de pedidos")
plt.ylabel("Probabilidade")
plt.legend()
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()

print(f"Probabilidade de exatamente 15 pedidos por hora: {probabilidade:.2%}")
```
